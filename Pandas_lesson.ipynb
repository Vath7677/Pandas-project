## Series (1D Labeled Array)

It is strictly one-dimensional, meaning it holds a single list of data. What makes it special is the Index—a built-in label for each item, allowing you to easily look up specific values.

```python
import pandas as pd

# Creating a Series from a list with custom index labels
temperatures = [22, 24, 19]
days = ['Monday', 'Tuesday', 'Wednesday']

temp_series = pd.Series(data=temperatures, index=days)
print(temp_series)
```

## DataFrames (2D Tabular Data)

A DataFrame is exactly like a full Excel spreadsheet or a database table. It is two-dimensional, meaning it has both rows and columns. In fact, you can think of a DataFrame as a collection of Series glued together side-by-side.

The Concept of Index vs. Columns:

- Index (Rows): The labels running vertically down the left side (identifying each row).
- Columns: The labels running horizontally across the top (identifying what data is in each vertical column).

```python
import pandas as pd

# Creating a DataFrame from a dictionary
data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'London', 'Paris']
}
df = pd.DataFrame(data)
print(df)
```

## The Core Concept

Think of Data I/O as the doors to your data workflow:
- Input (Reading): Bringing external data into Python so you can explore and change it. This converts your file into a DataFrame (a 2D table, just like a virtual spreadsheet).
- Output (Writing): Sending your cleaned or analyzed DataFrame back out of Python to save it as a file on your computer.

### Reading Data (Input)

These functions usually start with pd.read_... and are used to load a file into a DataFrame variable (commonly named df).
- pd.read_csv(): Reads plain text files separated by commas.
- pd.read_excel(): Reads standard Excel spreadsheets (.xlsx).
- pd.read_json(): Reads JSON files (common in web data).
- pd.read_sql(): Reads the results of a database query.

```python
df = pd.read_csv('products-100.csv')
```

### Writing Data (Output)

Once you are done modifying your data, you use methods attached directly to your DataFrame (starting with .to_...) to save it.
- .to_csv(): Saves your table as a new comma-separated file.
- .to_excel(): Saves your table as a new Excel file.
- index=False prevents pandas from writing the row numbers into the file

```python
df.to_excel("product-olala.xlsx", index=False)
```

## Viewing Data

These methods help you get a quick peek at the actual values inside your dataset.
- .head(n): Returns the first n rows of the dataset. If you leave n blank, it defaults to 5. Great for seeing what the top of your data looks like.
- .tail(n): Returns the last n rows. Defaults to 5. Useful for verifying that your entire file loaded correctly to the end.
- .sample(n): Returns a random selection of n rows. Defaults to 1.

```python
# df.head()
print(df.shape)
print(df.size)
df.head()
```

## Understanding Structure

These commands tell you about your data size, types, and labels rather than the data values themselves. (Note: attributes like shape, columns, and index do not use parentheses '()').
- .info(): Provides a concise summary of the dataset. It shows the number of rows/columns, the data type of each column (text, integer, decimal), and how many non-blank (non-null) values exist.
- .shape: Returns a tuple showing the dimensions of the data: (number_of_rows, number_of_columns).
- .columns: Returns a list of all the column names. Helpful when you have dozens of columns and need the exact spelling.
- .index: Returns information about the row labels (how the rows are numbered or named).

These methods give you a mathematical or categorical summary of the information in your columns.
- .describe(): Calculates quick statistical summaries for all numeric columns (count, mean, standard deviation, min, max, and percentiles).
- .value_counts(): Counts how many times each unique value appears in a specific column. Highly useful for categorical data (like seeing how many people are in each department).

## Column Selection: df['column'] vs. df.column

- df['column'] (Bracket style): This is the safest and most reliable method. It works for any column name, even if the name has spaces (e.g., df['First Name']) or matches a built-in Python function.
- df.column (Attribute style): This is a handy shortcut for typing quickly, but it fails if your column name has spaces, special characters, or shares a name with a pandas method (like .min or .sum).

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
        'Age': [25, 30, 35, 28],
        'City': ['NY', 'LA', 'Chicago', 'NY']}
df = pd.DataFrame(data)
df.index = ['row1', 'row2', 'row3', 'row4']
print(df)
```

## Label-based Indexing: .loc[]

- Explanation: Think of .loc as "locate by name." You use the exact text labels of your rows and columns to find data.
- Key rule: When you slice with .loc (e.g., A:C), it includes the last item.

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
        'Age': [25, 30, 35, 28],
        'City': ['NY', 'LA', 'Chicago', 'NY']}
df = pd.DataFrame(data)
df.index = ['row1', 'row2', 'row3', 'row4']
print(df)
print("\n")
print(df.loc[['row1','row3'],['Name','City']])
```

## Integer-based Indexing: .iloc[]

- Explanation: Think of .iloc as "locate by index (number)." You use the numerical position (0, 1, 2...) exactly like you would with standard Python lists.
- Key rule: When you slice with .iloc (e.g., 0:2), it goes up to but does not include the last number.

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
        'Age': [25, 30, 35, 28],
        'City': ['NY', 'LA', 'Chicago', 'NY']}
df = pd.DataFrame(data)
df.index = ['row1', 'row2', 'row3', 'row4']
print(df)
print("\n")
print(df.iloc[0:2, 0:1])
```

## Filtering (Boolean Conditions)

- Explanation: You can filter a DataFrame by passing true/false conditions inside the brackets [].
- Key rule: When combining multiple conditions, you must wrap each condition in parentheses () and use the bitwise operators & (and), | (or), and ~ (not). Standard Python and/or keywords will not work here.

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'Diana'],
        'Age': [25, 30, 35, 28],
        'City': ['NY', 'LA', 'Chicago', 'NY']}
df = pd.DataFrame(data)
df.index = ['row1', 'row2', 'row3', 'row4']
print(df)
print("\n")
older_than_28 = df['Age'] > 28
# print(df[older_than_28])
older_than_25_in_NY = (df['Age'] >= 25) & (df['City'] == 'NY')
# print(df[older_than_25_in_NY])
# older_than_25_or_in_NY = (df['Age'] > 28) | (df['City'] == 'LA')
# print(df[older_than_25_or_in_NY])
not_NY = ~(df['City']=='NY')
print(df[not_NY])
```
